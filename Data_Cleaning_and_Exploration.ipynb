```python
import pandas as pd
import random
```

```python
raw_data = pd.read_csv("202209-citibike-tripdata.csv")
```

```python
raw_data
```

```python
dropped_columns = raw_data.drop(columns=["ride_id", "start_station_id", "end_station_id"])
```

```python
dropped_nas = dropped_columns.dropna()
```

```python
dropped_dups = dropped_nas.drop_duplicates(ignore_index=True)
```

```python
data_set_length = len(dropped_dups)
```

```python
random_list = random.sample(range(0, data_set_length), 100000)
```

```python
sample_df = dropped_dups[dropped_dups.index.isin(random_list)]
sample_df
```

```python
sample_df.to_csv("random_sample_2022_09.csv")
```
