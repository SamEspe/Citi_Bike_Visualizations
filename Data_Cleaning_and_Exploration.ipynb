```python
import pandas as pd
import random
import datetime as dt
```

```python
raw_data = pd.read_csv("202209-citibike-tripdata.csv")
```

```python
dropped_columns = raw_data.drop(columns=["ride_id", "start_station_id", "end_station_id"])
```

```python
dropped_nas = dropped_columns.dropna()
```

```python
dropped_dups = dropped_nas.drop_duplicates(ignore_index=True)
```

```python
data_set_length = len(dropped_dups)
```

```python
random_list = random.sample(range(0, data_set_length), 100000)
```

```python
sample_df = dropped_dups[dropped_dups.index.isin(random_list)]
sample_df = sample_df.reset_index(drop=True)
```

```python
sample_df["start_lat"] = sample_df["start_lat"].round(3)
sample_df["start_lng"] = sample_df["start_lng"].round(3)
sample_df["end_lat"] = sample_df["end_lat"].round(3)
sample_df["end_lng"] = sample_df["end_lng"].round(3)
sample_df
```

```python
# Calculate the time each trip took, save it to a list

ride_time = []

for event in range(0,100000):
    # Convert started_at to datetime object, save it
    started_dt = dt.datetime.strptime(sample_df["started_at"][event], "%Y-%m-%d %H:%M:%S")
    # Convert ended_at to datetime object, save it
    ended_dt = dt.datetime.strptime(sample_df["ended_at"][event], "%Y-%m-%d %H:%M:%S")
    # Calculate elapsed time in seconds
    time_change = ended_dt - started_dt
    elapsed_seconds = time_change.seconds
    # Convert elapsed time in seconds to H:M:S string
    ride_length = str(dt.timedelta(seconds = elapsed_seconds))
    # Append string to series
    ride_time.append(ride_length)
```

```python
# Add series to dataframe as a new column
sample_df["ride_time"] = ride_time

sample_df
```

```python
sample_df.to_csv("random_sample_rounded_2022_09.csv")
```
